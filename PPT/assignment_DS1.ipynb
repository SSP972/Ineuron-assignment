9. Write a Python program to sort a list of integers in ascending order.

```python
def sort_list(nums):
    sorted_nums = nums.copy()
    for i in range(len(sorted_nums)):
        for j in range(i + 1, len(sorted_nums)):
            if sorted_nums[i] > sorted_nums[j]:
                sorted_nums[i], sorted_nums[j] = sorted_nums[j], sorted_nums[i]
    return sorted_nums

nums = [5, 2, 8, 1, 9]
sorted_nums = sort_list(nums)
print("Sorted list:", sorted_nums)
```

Output:
```
Sorted list: [1, 2, 5, 8, 9]
```

```python
def fibonacci(n):
    sequence = [0, 1]
    for i in range(2, n):
        next_term = sequence[i - 1] + sequence[i - 2]
        sequence.append(next_term)
    return sequence

n = 10
fib_sequence = fibonacci(n)
print("Fibonacci sequence up to", n, "terms:", fib_sequence)
```

15. Write a program to find the median of a list of numbers.

```python
def find_median(nums):
    sorted_nums = sorted(nums)
    n = len(sorted_nums)
    if n % 2 == 0:
        median = (sorted_nums[n // 2 - 1] + sorted_nums[n // 2]) / 2
    else:
        median = sorted_nums[n // 2]
    return median

nums = [5, 2, 8, 1, 9]
median = find_median(nums)
print("Median:", median)
```

Output:
```
Median: 5
```

23. Write a program to find the prime factors of a given number.

```python
def find_prime_factors(n):
    prime_factors = []
    divisor = 2
    while divisor <= n:
        if n % divisor == 0:
            prime_factors.append(divisor)
            n = n // divisor
        else:
            divisor += 1
    return prime_factors

num = 84
prime_factors = find_prime_factors(num)
print("Prime factors of", num, "are:", prime_factors)
```

Output:
```
Prime factors of 84 are: [2, 2, 3, 7]
```

26. Implement a function to find the mode of a list of numbers.

```python
from collections import Counter

def find_mode(nums):
    counts = Counter(nums)
    max_count = max(counts.values())
    modes = [num for num, count in counts.items() if count == max_count]
    return modes

nums = [1, 2, 3, 2, 1, 3, 4, 5, 4, 4]
modes = find_mode(nums)
print("Mode(s):", modes)
```

28. Implement a function to calculate the square root of a given number.

```python
def square_root(n):
    if n < 0:
        return None
    elif n == 0:
        return 0
    else:
        x = n
        y = (x + 1) // 2
        while y < x:
            x = y
            y = (x + n // x) // 2
        return x

num = 25
sqrt = square_root(num)
print("Square root of", num, "is", sqrt)
```

Output:
```
Square root of 25 is 5
```

30. Implement a function to find the minimum element in a rotated sorted list.

```python
def find_min_element(nums):
    left = 0
    right = len(nums) - 1
    while left < right:
        mid = (left + right) // 2
        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid
    return nums[left]

nums = [4, 5, 6, 7, 0, 1, 2]
min_element = find_min_element(nums)
print("Minimum element:", min_element)
```

Output:
```
Minimum element: 0
```

34. Implement a function to find the longest common prefix among a list of strings.

```python
def find_longest_common_prefix(strings):
    if not strings:
        return ""
    shortest_string = min(strings, key=len)
    for i, char in enumerate(shortest_string):
        for string in strings:
            if string[i] != char:
                return shortest_string[:i]
    return shortest_string

strings = ["flower", "flow", "flight"]
longest_common_prefix = find_longest_common_prefix(strings)
print("Longest common prefix:", longest_common_prefix)
```

Output:
```
Longest common prefix: fl
```

41. Write a Python program to find the smallest missing positive integer in a list.

```python
def find_missing_positive(nums):
    n = len(nums)
    for i in range(n):
        while 1 <= nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            nums[nums[i] - 1], nums[i] = nums[i], nums[nums[i] - 1]
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1
    return n + 1

nums = [3, 4, -1, 1]
missing_positive = find_missing_positive(nums)
print("Smallest missing positive integer:", missing_positive)
```

Output:
```
Smallest missing positive integer: 2
```

44. Implement a function to check if a given number is a perfect number.

```python
def is_perfect_number(n):
    if n <= 0:
        return False
    divisors = [1]
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n // i)
    return sum(divisors) == n

num = 28
if is_perfect_number(num):
    print(num, "is a perfect number.")
else:
    print(num, "is not a perfect number.")
```

Output:
```
28 is a perfect number.
```
