**Question 1: Check if a number is a power of three**

To determine if an integer is a power of three, we can repeatedly divide the number by three until it becomes 1 or not divisible by 3. If the final result is 1, then the original number is a power of three.

```python
def isPowerOfThree(n):
    if n <= 0:
        return False
    while n % 3 == 0:
        n //= 3
    return n == 1
```

**Question 2: Find the last number remaining after a specific algorithm**

To find the last number remaining, we can observe that after each iteration, the remaining number is reduced by half. If the current length is odd, the last number will remain the same. If the current length is even, the last number will be updated. We can use recursion to implement this algorithm efficiently.

```python
def lastRemaining(n):
    if n == 1:
        return 1
    if n % 2 == 0:
        return 2 * lastRemaining(n // 2)
    else:
        return 2 * lastRemaining(n // 2) + 1
```

**Question 3: Print all subsets of a set**

To print all subsets of a set, we can use a recursive approach where at each step, we have two choices: include the current element in the subset or exclude it. We can generate all subsets by recursively generating subsets for the remaining elements.

```python
def printSubsets(set):
    subsets = []
    generateSubsets(set, "", subsets)
    return subsets

def generateSubsets(set, current, subsets):
    if len(set) == 0:
        subsets.append(current)
        return
    generateSubsets(set[1:], current, subsets)
    generateSubsets(set[1:], current + set[0], subsets)
```

**Question 4: Calculate the length of a string using recursion**

To calculate the length of a string using recursion, we can check the base case where the string is empty. If the string is not empty, we can remove the first character and recursively calculate the length of the remaining substring.

```python
def stringLength(str):
    if str == "":
        return 0
    return 1 + stringLength(str[1:])
```

**Question 5: Count contiguous substrings with the same start and end character**

To count contiguous substrings with the same start and end character, we can iterate through each character in the string and expand around it to count substrings. We consider two cases: expanding around a single character and expanding around a pair of identical characters.

```python
def countSameStartEndSubstrings(S):
    count = 0
    for i in range(len(S)):
        count += 1  # Single character substring
        j = i - 1
        k = i + 1
        while j >= 0 and k < len(S) and S[j] == S[k]:
            count += 1
            j -= 1
            k += 1
        j = i
        k = i + 1
        while j >= 0 and k < len(S) and S[j] == S[k]:
            count += 1
            j -= 1
            k += 1
    return count
```

**Question 6: Tower of Hanoi**

To solve the Tower of Hanoi puzzle, we can use recursion. The main idea is to move N-1 disks from the source rod to the auxiliary rod, then move the Nth

 disk from the source rod to the destination rod, and finally move the N-1 disks from the auxiliary rod to the destination rod.

```python
def towerOfHanoi(N, source, destination, auxiliary):
    if N == 1:
        print("move disk 1 from rod", source, "to rod", destination)
        return 1
    moves = 0
    moves += towerOfHanoi(N-1, source, auxiliary, destination)
    print("move disk", N, "from rod", source, "to rod", destination)
    moves += 1
    moves += towerOfHanoi(N-1, auxiliary, destination, source)
    return moves
```

**Question 7: Print all permutations of a string**

To print all permutations of a string, we can use a recursive approach where at each step, we fix a character at the current position and recursively generate permutations for the remaining characters.

```python
def permutations(str):
    results = []
    generatePermutations(str, 0, len(str)-1, results)
    return results

def generatePermutations(str, left, right, results):
    if left == right:
        results.append(str)
    else:
        for i in range(left, right+1):
            str = swap(str, left, i)
            generatePermutations(str, left+1, right, results)
            str = swap(str, left, i)  # Backtrack

def swap(str, i, j):
    chars = list(str)
    chars[i], chars[j] = chars[j], chars[i]
    return "".join(chars)
```

**Question 8: Count total number of consonants in a string**

To count the total number of consonants in a string, we can iterate through each character and check if it is a consonant (not a vowel). We can use a set to store the vowels for efficient lookup.

```python
def countConsonants(str):
    vowels = {'a', 'e', 'i', 'o', 'u'}
    count = 0
    for char in str:
        if char.isalpha() and char.lower() not in vowels:
            count += 1
    return count
```
